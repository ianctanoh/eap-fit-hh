```python
import sys, os
from pathlib import Path

# find repo root (parent of "experiments") and add to sys.path
repo_root = Path(__file__).resolve().parents[1] if "__file__" in globals() else Path(os.getcwd()).parents[0]
sys.path.insert(0, str(repo_root))

from eap_fit_hh.inference import EKFTrainer
from eap_fit_hh.utils import compute_comp_xyz, distance, plot_cond_params, compute_eap, build_eap_M, build_axial_matrix, regroup_learned_params, get_surface_areas
from eap_fit_hh.C10_model import C10_model
```

```python
import jax.numpy as jnp
import matplotlib.pyplot as plt

import jaxley as jx

import jax.random as random
import tensorflow_probability.substrates.jax as tfp
tf = tfp.tf2jax
tfd = tfp.distributions
```

```python
C10 = C10_model()
cell = C10.build()
```

```python
ncomps = cell.shape[1]
```

```python
base_cell = C10.base_cell
base_cell_positions = compute_comp_xyz(base_cell)
```

```python
true_cell_positions = compute_comp_xyz(cell)
```

```python
# Grid dimensions
width = 288  # µm
height = 48  # µm
spacing = 6  # µm

# Create grid points
z = 50
y = jnp.arange(5, 5+width, spacing)
x = jnp.arange(-50, -50+height, spacing)
x_grid, y_grid = jnp.meshgrid(x, y)
n_points = len(y)*len(x)
# Stack into a set of (x, y, z) points
grid = jnp.column_stack((x_grid.ravel(), y_grid.ravel(),jnp.full(y_grid.size, z)))
```

```python
true_distances = distance(grid, true_cell_positions) # in um
true_distances_CM = true_distances * 10**(-4)
```

```python
# Create figure and 3D axis
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot compartments with lines
colors = plt.cm.tab10.colors
cell.vis(ax=ax, color=colors[0], dims=[0,1,2])

# Plot grid points
ax.scatter(grid[:, 0], grid[:, 1], grid[:, 2], c='r', marker='^', label='Electrodes')

# Labels and legend
ax.set_xlabel("X (μm)")
ax.set_ylabel("Y (μm)", labelpad=15)
ax.set_zlabel("Z (μm)")
ax.set_title("Inference of positions")
ax.legend()

x_limits = ax.get_xlim()
y_limits = ax.get_ylim()
z_limits = ax.get_zlim()

# Compute the range for each axis
x_range = x_limits[1] - x_limits[0]
y_range = y_limits[1] - y_limits[0]
z_range = z_limits[1] - z_limits[0]

# Find the max range
max_range = max(x_range, y_range, z_range)

# Set new limits so all axes have the same range
x_middle = sum(x_limits) / 2
y_middle = sum(y_limits) / 2
z_middle = sum(z_limits) / 2

ax.set_xlim([x_middle - max_range / 2, x_middle + max_range / 2])
ax.set_ylim([y_middle - max_range / 2, y_middle + max_range / 2])
ax.set_zlim([z_middle - max_range / 2, z_middle + max_range / 2])

# Labels and legend
ax.set_xlabel("X (μm)")
ax.set_ylabel("Y (μm)", labelpad=15)
ax.set_zlabel("Z (μm)")
ax.set_title("Cell Positions and Grid Points")
ax.legend()
```

```python
dt = 0.025  # ms
t_max = 30.0
times = jnp.arange(0, t_max+2*dt, dt)
n_timesteps = len(times)
current = jx.step_current(i_delay=0.0, i_dur=30., i_amp=.5, delta_t=dt, t_max=t_max) #1
cell.branch(0).comp(0).stimulate(current)

cell.record("v")

true_outputs = jx.integrate(cell)
```

```python
for i in range(ncomps):
    plt.plot(true_outputs[i,:])
```

```python
A = build_axial_matrix(cell)
padded_current = jnp.concatenate([jnp.array([0.]),current])
M = build_eap_M(A, true_distances_CM, 300)
v = true_outputs[:ncomps]
emissions_EAP = compute_eap(v, M, padded_current, 0, true_distances_CM, 300) + tfd.Normal(0,0.0001).sample((n_points, n_timesteps), seed=random.PRNGKey(0))
```

```python
plt.figure(figsize=(5.5, 5.))
im = plt.imshow(jnp.log(jnp.abs(M.T@M)), origin='upper', aspect='equal')  # top-left is [0,0]
cb = plt.colorbar(im)
cb.set_label(r'$\log(|\mathbf{H}^\top \mathbf{H}|)$')

# Put ticks at integer indices so it’s obvious which cell is which

# Draw a box around the top-left (row 0, col 0) cell
ax = plt.gca()

# Optional: label that cell

plt.xlabel('Column index')
plt.ylabel('Row index')
plt.tight_layout()
```

```python
HH_params = ['HHc_gNa', 'HHc_gK', 'HHc_gLeak']
params = {"soma": HH_params,
         "axon": HH_params,
         "oriprox": HH_params,
         "oridist": HH_params,
         "radprox": HH_params,
         "radmed": HH_params,
         "raddist": HH_params,
         "lm": HH_params}
```

```python
dynamics_std = {'v': 0.001, 'HHc_m': 0.0001, 'HHc_h': 0.0001, 'HHc_s': 0.0001, 'HHc_n': 0.0001,
                'i_HHc': 0.01}
```

```python
Trainer = EKFTrainer(cell = cell,
                     trainable_params = params,
                     dynamics_std = dynamics_std,
                     position_training = "rigid",
                     dt = dt,
                     grid = grid,
                     extracellular_resistivity = 300,
                     base_cell_positions = base_cell_positions)
```

```python
bounds_soma = {
        "HHc_gLeak": (1e-4, 4e-4),
        "HHc_gNa": (0.003, 0.015),
        "HHc_gK": (0.001, 0.005)
    }

bounds_axon = {
        "HHc_gLeak": (1e-6, 1e-5),
        "HHc_gNa": (0.08, 0.12),
        "HHc_gK": (0.015, 0.03)
    }

bounds_dend = {
        "HHc_gLeak": (1e-6, 1e-5),
        "HHc_gNa": (0.003, 0.015),
        "HHc_gK": (0.0005, 0.002)
    }

bounds = {"soma": bounds_soma,
         "axon": bounds_axon,
         "oriprox": bounds_dend,
         "oridist": bounds_dend,
         "radprox": bounds_dend,
         "radmed": bounds_dend,
         "raddist": bounds_dend,
         "lm": bounds_dend}

init_cond_params = Trainer.sample_cond_params(bounds = bounds, key = 0)
```

```python
init_pos_params = Trainer.sample_pos_params(radial_sd = 20, key = 0)
```

```python
def run(lr, init_key):
    init_cond_params = Trainer.sample_cond_params(bounds = bounds, key = init_key)
    init_pos_params = Trainer.sample_pos_params(radial_sd = 20, key = init_key) 
    out = Trainer.train(data = (emissions_EAP.T, current),
                init_cond_params = init_cond_params,
                init_pos_params = init_pos_params,
                obs_std = 0.0001,
                lr = lr,
                num_epochs = 100,
                output = ['cond_params', "cell_positions", "mll"])
    return out
```
