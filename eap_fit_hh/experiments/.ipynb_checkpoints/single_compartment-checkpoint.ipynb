```python
import sys, os
from pathlib import Path

# find repo root (parent of "experiments") and add to sys.path
repo_root = Path(__file__).resolve().parents[1] if "__file__" in globals() else Path(os.getcwd()).parents[0]
sys.path.insert(0, str(repo_root))

from eap_fit_hh.inference import EKFTrainer
from eap_fit_hh.utils import compute_comp_xyz, distance, get_surface_areas, plot_cond_params
```

```python
import jaxley as jx
from jaxley.channels import HH
import matplotlib.pyplot as plt

import jax.numpy as jnp
import jax.random as random
import tensorflow_probability.substrates.jax as tfp
tf = tfp.tf2jax
tfd = tfp.distributions
```

```python
ncomps = 1
comp = jx.Compartment()
branch = jx.Branch(comp, ncomp=ncomps)

branch.insert(HH())

branch.set("HH_gNa",  0.12)
branch.set("HH_gK", 0.02)
branch.set("HH_gLeak", 0.003)

branch.set("HH_eNa", 53.0)
branch.set("HH_eK", -107.0)
branch.set("HH_eLeak", -88.5188)
branch.set("v", -88.5188)
branch.set("radius",12.0)
branch.set("length",24.0)

cell_positions = compute_comp_xyz(branch)
surf_areas = get_surface_areas(branch)
surf_areas_CM2 = surf_areas * 10**(-8)

branch.init_states()
```

```python
n_points = 1
x = jnp.array([22.]) 
y = jnp.linspace(-6*(n_points-1)/2, 6*(n_points-1)/2, n_points) #all sites spaced by 6 um
z = jnp.array([0.])
grid = jnp.dstack(jnp.meshgrid(x, y, z)).reshape((n_points, 3))

distances = distance(grid, cell_positions) # in um
distances_CM = distances * 10**(-4)
```

```python
dt = 0.025  # ms
t_max = 20.0
times = jnp.arange(0, t_max+2*dt, dt)
n_timesteps = len(times)
current = jx.step_current(i_delay=0.0, i_dur=20., i_amp=2.5, delta_t=dt, t_max=t_max)
branch.comp(0).stimulate(current)

# Whatever is recorded will be returned by `jx.integrate()`. In this case,
# we record 8 compartments and voltage + 3 currents each. Therefore, `jx.integrate`
# will return a (8*4 x timesteps) array.
branch.comp("all").record("v")
branch.comp("all").record("HH_m")
branch.comp("all").record("i_HH")

true_outputs = jx.integrate(branch) 
```

```python
plt.plot(true_outputs[0,:])
plt.ylabel("V (mV)")
```

```python
def compute_eap(outputs):
    '''
    Input: Output of jx.integrate (of shape 3*T here). Last row is the HH_current.
    Output: (N_points, T) array of EAPs across time at each probe site.
    '''
    outputs = outputs.reshape(3, 1, -1)
    rho = 300
    resistivity = rho/(4.*jnp.pi)
    surf_currents = outputs[2:]
    currents = surf_currents * surf_areas_CM2[None,:,None]
    extr_voltage = jnp.sum(currents.sum(axis=0) / distances_CM[:,:,None], axis=1) * resistivity
    
    return extr_voltage
```

```python
emissions_EAP = compute_eap(true_outputs)+tfd.Normal(0,0.003).sample((n_points, n_timesteps), seed=random.PRNGKey(0))
plt.plot(emissions_EAP[0])
```

```python
trainable_params = ['HH_gNa', 'HH_gK', 'HH_gLeak']
```

```python
dynamics_std = {'v': 0.0001, 'HH_n': 0.00001, 'HH_m': 0.00001, 'HH_h': 0.00001, 'i_HH': 0.01}
```

```python
Trainer = EKFTrainer(cell = branch,
                     trainable_params = trainable_params,
                     dynamics_std = dynamics_std,
                     dt = dt,
                     grid = grid,
                     extracellular_resistivity = 300,
                     obs_std = 0.003)
```

```python
init_bounds = {"HH_gLeak": (0.0001, 0.003),
          "HH_gNa": (0.05, 0.2),
          "HH_gK": (0.005, 0.05)}

init_params = Trainer.sample_cond_params(bounds = init_bounds, key = 2) #save 2
```

```python
out = Trainer.train(data = (emissions_EAP.T, current),
            init_cond_params = init_params,
            lr = 0.004,
            num_epochs = 3000)
```

```python
inferred_params, true_params = out["cond_params"], Trainer.ref_cond_params
```

```python
plot_cond_params(params = inferred_params, true_params = true_params, 
                 label_params = {"HH_gLeak": 'g_L',"HH_gNa": "g_Na", "HH_gK": "g_K"})
```
